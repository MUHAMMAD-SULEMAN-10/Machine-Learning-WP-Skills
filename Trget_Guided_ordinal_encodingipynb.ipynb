# Target Guided Ordinal Encoding

It is a technique used to encode categorical varaiables based on their relationship with the target variable. This encoding technique is useful when we have a categorical varaiable with a large number od unique categories,  and we want to use this variable as a feature inour machine leaninig moadel.


In Target Guided Ordinal Encoding, we replace each category in the categorical varaible with a numerical value based on the mean or median of the target varaible for that category. This creates a monotronic relationship between the categorical variable and the target variable, which can improve the predictive power of our model.

```python
import pandas as pd

# create a sample dataframe with a categorical variable and a target variable

df = pd.DataFrame({
    'city': ['New York', 'London', 'Paris', 'Tokyo', 'New York', 'paris'],
    'price': [200, 150, 300, 250, 180, 320]
})
```

```python
df
```

```python
## calculate the mean proce for each city

df.groupby('city')['price']
```

```python
df.groupby('city')['price'].mean() # take mean when there is no outlieers and take medain when  there are outliers


# groupby is used to split data into groups based on some feature(s) and then apply aggregation or transformation on each group.
# df.groupby('city') → Groups the data by unique values in the city column.
```

```python
mean_price= df.groupby('city')['price'].mean().to_dict ()
mean_price
```

```python
## repalce each city with its mean price

df['city_encoded'] = df['city'].map(mean_price)
```

```python
df
```

## now we can do it with any dataset like we have a categorical feature and its dependencies has on the target feature and then we can realte it with the target feature

```python
import seaborn as sns
df=sns.load_dataset('tips')
```

```python
df
```

```python
mode_day = df.groupby('total_bill')['day'].apply(lambda x: x.mode()).to_dict ()
mode_day
```

```python
mode_day = df.groupby('total_bill')['day'].agg(lambda x: x.mode()[0])
```

```python
mode_day
```

```python

df['bill_encoded'] = df['total_bill'].map(mode_day)
```
