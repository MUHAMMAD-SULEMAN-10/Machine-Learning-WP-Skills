#Normalization-Min Max Scalar

it is mostly used in deep learning techniques

```python
import seaborn as sns
```

```python
df = sns.load_dataset('taxis')
```

```python
df
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
min_max = MinMaxScaler()
```

```python
min_max.fit(df[['distance', 'fare', 'tip']])
```

```python
min_max.transform(df[['distance', 'fare', 'tip']])
```

```python
# we can do in one line

min_max.fit_transform(df[['distance', 'fare', 'tip']])
```

```python
# and now if we want to scale new data points then these will also scaled but make sure pass values in a double brackets

min_max.transform([[1,3,4]])
```
