#What is Imbalanced Data in Machine Learning?

Imbalanced data means your dataset has unequal distribution of classes.

Example:
Suppose you are doing fraud detection:

98% transactions → Not Fraud

2% transactions → Fraud

Here, the model sees too many “Not Fraud” examples and very few “Fraud” ones.

❗ Why is this a problem?

Because the model learns to predict the majority class every time.

Example:
If the model predicts “Not Fraud” for everything → accuracy becomes 98%, but still useless.

So, imbalanced data leads to:

Misleading accuracy

Poor recall/precision on minority class

Biased model performance



# What is Up Sampling?

Upsampling means increasing the number of samples of the minority class.

✔ How it works?

You duplicate or synthetically generate more minority class samples.

Example:
Fraud (minority = 200 samples) → increase to 2000 samples.

✔ Why use Upsampling?

To balance the class distribution

To improve model learning for minority class

To avoid bias toward majority class

Common Upsampling Techniques:

Random Oversampling (simple duplication)

SMOTE (Synthetic Minority Oversampling Technique)

ADASYN (Adaptive Synthetic)

# What is Down Sampling?

Downsampling means reducing the number of samples in the majority class.

✔ How it works?

You randomly remove samples from the majority class.

Example:
Non-Fraud (majority = 10,000 samples) → reduce to 2,000.

✔ Why use Downsampling?

To reduce class imbalance

To make model treat both classes fairly

Dataset becomes smaller → faster training

Disadvantages:

Loss of useful information

Might remove important patterns from majority class

##Summary

Concept	Meaning	Why Do It?

Imbalanced Data	One class has much more data than the other	Causes biased/incorrect predictions

Upsampling	Increase minority class samples	Helps model learn minority class better

Downsampling	Reduce majority class samples	Makes dataset balanced; faster training

```python
import numpy as np
import pandas as pd
```

```python
from types import new_class
# Set the random seed for reproducibility
np.random.seed(123) # data generate from numpy and it will not change seed data

#Create a dataframe with two classes
n_samples = 1000
class_0_ratio = 0.9
n_class_0 = int(n_samples * class_0_ratio) # class 0 has 900 data points
n_class_1 = n_samples - n_class_0 # now remaining 100
```

```python
# We want basically an imbalanced data set

n_class_0 , n_class_1
```

```python
class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc =0, scale=1, size=n_class_0),
    'target': [0]*n_class_0 # has nine husndred 000...

})

# here mean = 0 and standrad deviation = 1 and n_class_0 = 900 size


class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc =2, scale=1, size=n_class_1),
    'target': [1]*n_class_1

})
```

```python
df=pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df
```

```python
df.head()
```

```python
df['target'].value_counts()
```

##Upsampling

```python
df_minority =  df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
df_minority.head()
```

```python
df_majority.head()
```

```python
# Upsampling perform

from sklearn.utils import resample # it is library


```

```python
df_minority_upsample = resample(df_minority,
                                replace=True, ##Sample with replacement and create new datapoints
                                n_samples=len(df_majority), # to match the majority class
                                random_state=42

                                )
```

```python
df_minority_upsample.shape
```

```python
df_minority_upsample['target'].value_counts()
```

```python
df_upsampled = pd.concat([df_majority, df_minority_upsample])
```

```python
df_upsampled['target'].value_counts()
```

## DownSampling

```python


class_0 = pd.DataFrame({
    'feature_1': np.random.normal(loc=0, scale=1, size=n_class_0),
    'feature_2': np.random.normal(loc =0, scale=1, size=n_class_0),
    'target': [0]*n_class_0 # has nine husndred 000...

})

# here mean = 0 and standrad deviation = 1 and n_class_0 = 900 size


class_1 = pd.DataFrame({
    'feature_1': np.random.normal(loc=2, scale=1, size=n_class_1),
    'feature_2': np.random.normal(loc =2, scale=1, size=n_class_1),
    'target': [1]*n_class_1

})
```

```python
df=pd.concat([class_0, class_1]).reset_index(drop = True)
```

```python
df_minority =  df[df['target']==1]
df_majority = df[df['target']==0]
```

```python
df_majority_downsample = resample(df_majority,
                                replace = False, ##Sampling witout replacement and not create new datapoints
                                n_samples=len(df_minority), # to match the majority class
                                random_state=42

                                )
```

```python
df_majority_downsample.shape
```

```python
df_downsample = pd.concat([df_minority, df_majority_downsample])
```
