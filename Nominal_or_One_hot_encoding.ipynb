# Nominal One Hot Encoding Techniques

Nominal encoding is a technique used to transoform categorical variabls that have no intrinsic ordering into numerical values that can be used in machine learninig models. One commone method for nominal encodign is one-hot encoding, whic creates a binary vector for each category in the variable.


###Disadvantages of Nominal One-Hot Encoding

High dimensionality
Creates many columns if a feature has many categories (curse of dimensionality).

Increased memory usage
More columns â†’ more storage and computation cost.

Slower model training
Large number of features increases training time.

Sparse data problem
Most values are 0, making data sparse and inefficient.

Not suitable for high-cardinality features
Columns like City, Product ID can create thousands of features.

Harder to interpret
Too many dummy variables reduce model interpretability.

One-line exam answer

One-Hot Encoding increases dimensionality and computational cost, especially for features with many categories.

#Nominal or one hot encoding

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
# create a sample dataframe with categorical variable

df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'red', 'blue']
})

```

```python
df
```

```python
## create an instance of One hot encoder

encoder = OneHotEncoder()
```

```python
encoded = encoder.fit_transform(df[['color']])
```

```python
encoded
```

```python
#convert it into an array

encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
encoded
```

```python
encoder.get_feature_names_out()
```

```python
import pandas as pd
encoded_df =pd.DataFrame(encoded, columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
pd.concat([df, encoded_df], axis = 1)

# now our color is also concat
```
