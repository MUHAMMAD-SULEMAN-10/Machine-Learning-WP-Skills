# Standardization

```python
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df['total_bill']
```

```python
total_bill = list(df['total_bill'])
```

```python
total_bill
```

```python
sns.histplot(total_bill)
```

```python
import numpy as np
mean  = np.mean(total_bill)
std = np.std(total_bill)
```

```python
mean,std
```

```python
## apply z - score formula

normalized_data = []
for i in total_bill:
  z_score=(i-mean)/std
  normalized_data.append(z_score)
```

```python
print(normalized_data)
```

```python
sns.histplot(normalized_data)

#now data is scaled down in a same range
```

# use standarad scalar

```python
from sklearn.preprocessing import StandardScaler


# Now Understand the process here the automatically the Z - Score will calculated and fit() function will calculate the mean and standard deviaiton.
# but the values of mean and standard deviaiton apply on the dataset is the transform.
```

```python
scaler = StandardScaler()
```

```python
scaler

#it will show StandardScaler() object
```

```python
df.head()
```

```python
scaler.fit(df[['total_bill', 'tip']])

# give your feature in a double brackets
```

```python
scaler.transform(df[['total_bill', 'tip']])
```

```python
# now if you want to store in the DataFrame
import pandas as pd
pd.DataFrame(scaler.transform(df[['total_bill', 'tip']]),columns=['total_bill', 'tip'])

#Columns are added again because StandardScaler.transform() returns a NumPy array without column names.
```
